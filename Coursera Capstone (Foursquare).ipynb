```python
import numpy as np
import pandas as pd
```

Create dataframe from Wikipedia table about Neighborhoods in Toronto

```python
df_li = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")
df = df_li[0]
df = df.loc[df["Borough"] != "Not assigned"]
df
```

Check if any Postal Codes are duplicates:

```python
df.duplicated(subset="Postal Code").value_counts()
```

0 Postal Codes are duplicates

Check if any neighborhoods have the value "Not assigned"

```python
df.loc[df["Neighborhood"] == "Not assigned"]
```

0 neighborhoods have the value "Not assigned"

Get Toronot Neighborhood locations

```python
df_coord = pd.read_csv("http://cocl.us/Geospatial_data")
df_coord.head()
```

Merge Neighborhood df with coordinates

```python
df = pd.merge(df,df_coord, on="Postal Code")
df
```
